**任务3.2：**分析电影净利润（票房-成本）随着年份变化的情况，并简单进行分析。

```python
earned = clean_data['revenue_adj'] - clean_data['budget_adj']
clean_data['earned'] = earned
earned_plt = pd.DataFrame(clean_data.groupby(['release_year'])['earned'].agg('mean'))
earned_plt = earned_plt.reset_index('release_year')
sb.lineplot(data = earned_plt, x='release_year', y='earned');
```

从上图可以看出，整个电影的净利润走向并不是一个比较平滑的折线。从最初到前期的起伏十分巨大，说明此时电影行业的利润情况是不稳定的。
在折线的最后我们可以发现，整个利润逐渐趋向于一个平稳的状态。说明电影行业发展到后面，利润情况已经难有巨大变化，预算和票房已经难有较大差值。
